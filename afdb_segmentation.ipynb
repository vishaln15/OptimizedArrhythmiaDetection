```python
import pandas as pd
import numpy as np
import os
import functools
import operator
```

```python
lengthInSeconds = 5
strideInSeconds = 1
fs = 250
length = lengthInSeconds * fs
stride = strideInSeconds * fs
```

```python
onemindf = pd.read_csv('hugedfrpeak.csv', index_col=[0])
```

```python
onemindf
```

```python
def strided_app(a, L, S ):  
    nrows = ((a.size-L)//S)+1
#     print(nrows)
    n = a.strides[0]
    return pd.DataFrame(np.lib.stride_tricks.as_strided(a, shape=(nrows,L), strides=(S*n,n)))
```

```python
L_arr = [length]*len(onemindf)
S_arr = [stride]*len(onemindf)
labels = onemindf['af'].copy()
del onemindf['af']
```

```python
final_ecg_list = list(map(strided_app, onemindf.to_numpy(), L_arr, S_arr))
final_label_list = list(map(lambda x: [x]*56, labels.values))
```

```python
final_df = pd.concat(final_ecg_list)
```

```python
final_df
```

```python
final_labels_list_flat = functools.reduce(operator.iconcat, final_label_list, [])
```

```python
final_df.insert(0,'af', final_labels_list_flat)
```

```python
final_df.reset_index(inplace=True, drop = True)
```

```python
final_df
```

```python
nan_df = final_df[final_df.isnull().any(axis=1)]
```

```python
final_df.drop(nan_df.index, inplace=True)
```
